```python
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
glyphs :
    a plot consists of geometric shapes either in the form of a line, circle, etc. 
    So, Glyphs are nothing but visual shapes that are drawn to represent the data such as circles, squares, lines,rectangles...
    

        circle : Represents data points as circles on the plot.
        line: Connects data points with lines to visualize trends or relationships.
        square: Represents data points as squares.
        bar: Represents data as bars, often used for categorical data.
        patch: Represents a closed polygon defined by a set of coordinates.
        wedge: Represents data as wedges (partial circles) for creating pie charts.
        image: Displays images on the plot.
        text:  Allows adding text to the plot at specified coordinates.
```

```python
data = px.data.stocks()
data
```

```python
# ex we draw line and circle

from bokeh.plotting import figure, output_file, show
output_file("first.html")
fig = figure(title="My_first")
cr = fig.circle(data["GOOG"],data["AAPL"])
glyph = cr.glyph
glyph.size = 35
glyph.fill_color = "red"

show(fig)
```

```python
from bokeh.plotting import figure, output_file, show
output_file("first.html")
fig = figure(title="My_first")
line = fig.line(data["GOOG"],data["AAPL"])
glyph = line.glyph
glyph.line_color = "red"
glyph.line_width = 2
show(fig)
```

```python
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
from bokeh.plotting import figure, output_file, show
output_file("first.html")
fig = figure(title="My_first")
fig.xaxis.axis_label = "x-label"
fig.yaxis.axis_label = "y-label"
fig.scatter(data["GOOG"],data["AAPL"] , legend_label = "yehhe" )
fig.legend.title = "hello world"
fig.legend.location = "top_left"

show(fig)
```

```python
Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in 
real time?
```

Bokeh is an open-source Python library used for interactive data visualization. 
Bokeh allows you to create interactive and interactive web-based plots, dashboards, and applications for 
displaying data in a web browser.

The Bokeh server is a key feature of the Bokeh library that enables the creation of interactive web applications and dashboards.
With the Bokeh server, you can build dynamic visualizations that respond to user interactions in real-time.

```python
import numpy as np
```

```python
import bokeh.io
import bokeh.plotting
import numpy as np
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
from bokeh.models import Div, Spinner, TextInput
from bokeh.layouts import layout


output_file("hero.html")

x= np.arange(0,10,1)
y = np.random.randint(5,15 , 10)

fig = figure(title= "hello world" , x_axis_label = "X" , y_axis_label = "Y")
points = fig.circle(x , y)

div = Div(text="<p>Enter a size : </p>")
spinner = Spinner(title= "Circle size" , low =5 , high = 10,step=1 , value=points.glyph.size , width=200)

spinner.js_link("value", points.glyph,"size")

text_input = TextInput(value=points.glyph.fill_color , width=200)
text_input.js_link("value" , points.glyph , "fill_color")

layout = layout([[div,spinner] ,[text_input] , [fig]])

show(layout)
```

```python
Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?
```

```python
from flask import Flask
from flask import render_template


app = Flask("__name__")


@app.route("/")
def my_bokeh():
    import bokeh.io
    import bokeh.plotting
    bokeh.io.output_notebook()
    from bokeh.plotting import figure, output_file, show

    output_file("flask_bokeh")
    
    x = [1,2,3,4,5,6,7,8,9]
    y = [9,8,7,6,5,4,3,2,1]
    
    p = figure(title = "using flask" , x_axis_label = "X", y_axis_label = "Y")
    fig = p.line(x,y)
    show(fig)
    
    return 



if __name__ =="__main__":
    app.run(host= "0.0.0.0")
```
