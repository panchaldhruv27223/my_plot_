```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [223,456,770,120,222,444,200,300]

plt.plot(product_id , online_price)

plt.xlabel("product")
plt.ylabel("price")
plt.title("product - price")

plt.show()
```

```python
plt.plot(product_id , online_price , color= "red")
#  you cane change color
```

```python
plt.plot(product_id , online_price , color= "red" , marker = "+")
#  for this pattern passs + 
```

```python
plt.plot(product_id , online_price , color= "red" , marker  = "D")
#  for this pattern pass D 
```

```python
plt.plot(product_id , online_price , color= "red" , marker = "D" , markersize = 10)
```

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [223,456,770,120,222,444,200,300]
offline_price  = [400 , 300 , 100 , 333, 444,565, 899 , 900]
chor_bazar = [100,200,300,400,150,250,450,500]
```

```python

plt.plot(product_id , online_price , color = "blue", markersize = 10)
plt.plot(product_id , offline_price , color = "red" ,marker = "D" ,markersize = 10)
plt.plot(product_id , chor_bazar, color = "green" ,marker = "+" ,markersize = 10)
plt.xlabel("product_")
plt.ylabel("price")
plt.title("product -- price ")
plt.show()
```

```python
plt.plot(product_id , online_price , color = "blue", markersize = 10 , label = "online_price")
plt.plot(product_id , offline_price , color = "red" ,marker = "D" ,markersize = 10 , label = "offline_price")
plt.plot(product_id , chor_bazar, color = "green" ,marker = "+" ,markersize = 10 , label = "chor_bazar")
plt.xlabel("product_")
plt.ylabel("price")
plt.title("product -- price ")
plt.legend()
plt.show()
```
