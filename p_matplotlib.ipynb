```python
data={
    "name" : ["dhruv","pravin","chintan","parth","hiten","ved","jay","dhruv","pravin","chintan","parth","hiten","ved","jay"],
    "salary" : [10,20,5,6,7,9,90,33,55,44,66,77,60,10]
}
print(len(data["name"]))
print(len(data["salary"]))
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df_new = df.groupby("name").sum()
```

```python
df_new["salary"].plot(kind="pie",
                     figsize=(5, 6),
                    autopct='%.1f%%', # add in percentages
                    startangle=0,
                    shadow=True)
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink',"red"]

df_new["salary"].plot(kind="pie",
                     figsize=(15, 6),
                    autopct='%.1f%%', # add in percentages
                    startangle=90,
                    shadow=True,        # turn off labels on pie chart
                    pctdistance=.50,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                    colors=colors_list,  # add custom colors
                    explode = [0,0,0,0,1,0,1]  ### fo extract it is use
                     )
plt.axis('equal')

plt.title("this is my pie chart" , y = 1.2)
plt.legend(loc="upper left",labels=df_new.index)
```

```python
fig = plt.figure(1,figsize=(5,5))
plt.plot(x, y, marker="o", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="data set-1")
plt.plot(x, y+500, marker="*", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="data set-1")

plt.legend(loc="lower right")
plt.show()
```

```python
fig = plt.figure(1,figsize=(5,5))
plt.plot(x, y, marker="o", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="data set-1")
plt.semilogy(x, y+500, marker="*", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="data set-1")

plt.legend(loc="lower right")
plt.show()
```

```python
fig = plt.figure(1,figsize=(5,5))
plt.semilogx(x, y, marker="o", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="data set-1")
plt.semilogy(x, y+500, marker="*", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="data set-1")

plt.legend(loc="lower right")
plt.show()
```

```python
fig = plt.figure(1,figsize=(5,5))
plt.semilogx(x, y, marker="o", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="$y = x^2$")
plt.semilogy(x, y+500, marker="*", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="$y = x^2 +500 $")

plt.legend(loc="lower right")
plt.show()
```

```python
fig = plt.figure(1,figsize=(5,5))
plt.semilogx(x, y, marker="o", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="$y = x^2$")
plt.semilogy(x, y+500, marker="*", markersize=10, mfc="white", linestyle="--", alpha=0.80, label="$y = x^2 +500 $")
plt.legend(loc="lower right")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
