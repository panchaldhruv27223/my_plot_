```python
The subplot() function is used in various programming libraries and environments for 
creating multiple plots or subplots within a single figure. It allows you to organize 
multiple plots in a grid-like arrangement, which is helpful for comparing different datasets, 
visualizing multiple aspects of data, or presenting related information side by side.
```

```python
# plt-1
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.title("plt-1")

plt.plot(x,y)

# plt2
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.title("plt-2")

plt.plot(x1,y1 , c="green")

# plt3
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.title("plt-3")

plt.plot(x2,y2 ,c="red")
# plt.plot(x2,y2 ,"bo")

# plt4
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.title("plt-4" )

plt.plot(x3,y3, c="yellow")
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
A bar plot, also known as a bar chart, is a graphical representation of data that uses rectangular bars to show 
the values of different categories or groups.
Bar plots are typically used to display categorical data, where the different categories are discrete and distinct.

It is important to note that while bar plots are effective for certain types of data, 
they may not be suitable for all situations. For continuous data, where values fall along a range, other types of plots 
such as line plots or histograms might be more appropriate. Choosing the right type of plot depends on the nature of your 
data and the specific insights you want to convey.

use :
    Comparison of data
    Visualizing distributions
    Handling discrete data
    Handling large datasets
    Visualizing trends over time
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("Bar -- plot")
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel("profit")
plt.ylabel("company")
plt.title("Bar -- plot")
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a 
graphical summary of the distribution of a dataset. 
It presents a concise and informative representation of the data'smedian, quartiles, and potential outliers. 
The plot consists of a rectangular "box" and lines("whiskers") extending from it.

use :
    Visualizing data distribution
    Comparison of distributions
    Detection of outliers
    Assessment of data symmetry and skewness
    Handling large datasets
    
    -Box plots are widely used in exploratory data analysis, statistical analysis, and data visualization 
     due to their ability to provide valuable insights into the distribution and characteristics of the data,
     especially when comparing multiple datasets or looking for patterns and outliers.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# plt.boxplot([box1,box2],vert=False)
plt.boxplot([box1,box2],labels=["d","h"])
```
