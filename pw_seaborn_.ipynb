```python
Q.1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
Seaborn is a powerful data visualization library built on top of Matplotlib, 
designed to make visualizing data more straightforward and aesthetically pleasing. 

sns.scatterplot :
     scatter plots show the relationship between two numerical variables, each represented by a point on the graph.
     It helps visualize the correlation or pattern between the two variables.

sns.barplot :
     Bar plots are used to compare categorical data, typically represented on the x-axis, while the height of the bars
     represents the corresponding values on the y-axis. 
     They are useful for visualizing counts or aggregations of data.

sns.boxplot :
     Box plots, also known as box-and-whisker plots, provide a summary of the distribution of numerical data. 
     They show the median, quartiles, and identify potential outliers in the data.

sns.histplot :
    Histograms display the distribution of a single numerical variable by dividing it into bins and showing the frequency
    or count of data points falling into each bin.

sns.jointplot :
    jointplot is used to visualize the relationship between two numerical variables by plotting their joint distribution 
    along with individual univariate distributions on the x and y axes. It is particularly useful for exploring the 
    correlation and distribution of two variables simultaneously.

sns.pairplot :
     pairplot is used to create a grid of scatter plots to visualize the pairwise relationships between multiple numerical 
     variables in a dataset. 
     It is an excellent tool for understanding correlations and identifying patterns in data with multiple dimensions.
        
sns.heatmap :
    Heatmaps are used to visualize the matrix-like data, where each cell is color represents the value at the intersection
    of a row and column. They are effective for displaying correlations, pairwise relationships, or matrices.
```

Q.6 Use the "flights" dataset from seaborn to plot a heatmap.

```python
f = sns.load_dataset("flights")
f
```

```python
flights = sns.load_dataset("flights").pivot("month","year","passengers")
flights
```

```python
import pandas as pd
import numpy as np
import datetime
```

```python
sns.heatmap(flights , linewidths=0.1,cmap="crest" , vmin=(np.min(flights)).min() , vmax=(np.max(flights)).max() )
```

```python
(np.min(flights)).min()
```

```python
sns.heatmap(flights , linewidths=0.1, cmap="crest",annot=True, fmt=".0f", vmin=(np.min(flights)).min() , vmax=(np.max(flights)).max() )
```
